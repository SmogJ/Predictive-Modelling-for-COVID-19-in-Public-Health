```python
def get_df(csv_folder):
    
    # Create a DataFrame of all csv files in target tables
    combined_df= pd.DataFrame()
    
    # Getting the list folders
    csv_folders= sorted(os.listdir(csv_folder))
    folders_path=[]
    
    for folder in csv_folders[1:]:
        folders_path.append(csv_folder + "/" + folder)
        
    for folder in folders_path:
        # file_paths= os.listdir(folder)
        file_paths= glob.glob(folder + "/*.csv")
        
        for file in file_paths:
            if file.endswith(".csv"):
                df= pd.read_csv(file)
                combined_df= pd.concat([combined_df, df], axis=0, ignore_index= False)
                
    gc.collect()
    return combined_df
```
